Scholarship Eligibility Checker and Visualizer
----------------------------------------------

This Python script collects information about a group of students and checks their
eligibility for a scholarship based on age and grade level.

Eligibility Criteria:
- Age must be 18 or older
- Grade must be 12 or higher

The script also generates a color-coded bar chart to visualize each student's age,
highlighting eligibility with green bars and ineligibility with red bars.

Features:
- User input for any number of students
- Automatic eligibility checking with detailed reasons
- Visual chart with age threshold line and annotations
- Styled using Matplotlib

Dependencies:
- matplotlib


```python
# Scholarship Eligibility Checker

# Define how many students to input
num_students = int(input("Enter number of students: "))

students = []

# Collect student information using input
for i in range(num_students):
    print(f"\nEnter details for student {i + 1}:")
    name = input("Name: ")
    age = int(input("Age: "))
    grade = int(input("Grade: "))
    
    student = {"name": name, "age": age, "grade": grade}
    students.append(student)

print("\n=== üéì Scholarship Eligibility Results ===\n")

# Check eligibility
for student in students:
    name = student["name"]
    age = student["age"]
    grade = student["grade"]

    if age >= 18 and grade >= 12:
        print(f"\033[92m‚úÖ {name} is eligible! üéâ Age: {age}, Grade: {grade}\033[0m")
    else:
        reason = []
        if age < 18:
            reason.append(f"age is {age} (must be >= 18)")
        if grade < 12:
            reason.append(f"grade is {grade} (must be Grade 12)")
        reason_text = " and ".join(reason)
        print(f"\033[91m‚ùå {name} is NOT eligible. Reason: {reason_text}.\033[0m")
```

```python
import matplotlib.pyplot as plt

# Prepare data
names = [student["name"] for student in students]
ages = [student["age"] for student in students]
colors = ['green' if age >= 18 else 'red' for age in ages]

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(names, ages, color=colors)
plt.axhline(y=18, color='blue', linestyle='--', label="Eligibility Age Threshold")
plt.title("Student Age Eligibility for Scholarship")
plt.xlabel("Student")
plt.ylabel("Age")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
